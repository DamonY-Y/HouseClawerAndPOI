# 1.获取百度坐标

```python
import requests
import json
import pandas as pd
from time import sleep
```

```python
data=pd.read_csv("addr_df.csv") #读取数据
address = data['0']
print(address)
```

```python
bdlng = []   # 存放经度
bdlat = []    # 存放纬度
for i in range(len(address)):
    url = 'http://api.map.baidu.com/geocoding/v3/?address='
    url_last = '&output=json&ak=CCveNaEXtUnRMuipcNXb6muXase7rYzz&callback=showLocation'
    url = url + '金牛区'+address[i] + url_last
    print(address[i])
    print(url)
    r = requests.get(url).text
    r = r.replace('showLocation&&showLocation(', '').replace(')', '')
    
    info = json.loads(r)
    
    bdlng.append(info['result']['location']['lng'])
    bdlat.append(info['result']['location']['lat'])
#     sleep(1)
print(bdlng)
print(bdlat)
dic = {'小区名称':address,'百度经度':bdlng,'百度纬度':bdlat}
data_crood = pd.DataFrame(dic)
print(data_crood)
```

```python
data_crood.to_csv('address_crood.csv')
print('ok')
```

```python
data_crood.to_excel('address_crood.xlsx')
```

# 2.利用百度API获取小区坐标周边配套

```python
import pandas as pd #导入数据分析模块
import requests #导入网页请求模块
```

```python
data=pd.read_csv("test.xlsx") #读取数据
lat_partion=data["百度纬度"] #纬度
lng_partion=data["百度经度"] #经度
store=data["项目名称"] #门店
targets=[]  #用来存要查找的名称
addresss=[]  #用来存放地址
stores=[]     #用来存放项目名称
distances=[]  #用来存放距离
```

```python
keyword = '菜市场'
radius = '5000'

for i in range(len(lat_partion)):
        location=str(lat_partion[i])+","+ str(lng_partion[i]) #构造圆中心点的经纬度
#         print(location) #测试            
        url="http://api.map.baidu.com/place/v2/search?query="+keyword+"&page_size=10&page_num=0&location="+location+"&radius="+str(radius)+"&output=json&ak=CCveNaEXtUnRMuipcNXb6muXase7rYzz&scope=2"#构造请求网址
        print(url) #测试请求接口拼接是否正确，此url可以直接复制到浏览器查看返回结果       
```

```python
response=requests.get(url) #发出请求
answer=response.json() #返回结果json化
# print(answer) #测试
print("一共%s条数据"%len(answer['results']))
for k in range(len(answer['results'])):  
    target=answer['results'][k]['name'] #标的物名称
    address=answer['results'][k]['address'] #地址
    distance=answer['results'][k]['detail_info']['distance'] #距离
    targets.append(target)
    addresss.append(address)
    stores.append(store[i])
    distances.append(distance)
    print(target,distance,address)
```

```python
def get_directory(keyword,radius): #定义圆形区域检索函数
    for i in range(len(lat_partion)):
        location=str(lat_partion[i])+","+ str(lng_partion[i]) #构造圆中心点的经纬度
        #print(location) #测试
        try:                
            url="http://api.map.baidu.com/place/v2/search?query="+keyword+"&page_size=10&page_num=0&location="+location+"&radius="+str(radius)+"&output=json&ak=CCveNaEXtUnRMuipcNXb6muXase7rYzz&scope=2"#构造请求网址
            #print(url) #测试请求接口拼接是否正确，此url可以直接复制到浏览器查看返回结果
            response=requests.get(url) #发出请求
            answer=response.json() #返回结果json化
            
            print("一共%s条数据"%len(answer['results']))
            
            for k in range(len(answer['results'])):  
                target=answer['results'][k]['name'] #标的物名称
                address=answer['results'][k]['address'] #地址
                distance=answer['results'][k]['detail_info']['distance'] #距离
                targets.append(target)
                addresss.append(address)
                stores.append(store[i])
                distances.append(distance)
                print(target,distance,address)
                
        except:
            print("the circle contains no message")
            stores.append(store[i])
            targets.append("")
            distances.append("")
            addresss.append("")
           
```

```python
keyword=['学校','公园',] #输入poi的关键词
radius = 50000  #输入半径
get_directory(keyword,radius)  #调用函数get_directory
dict={"项目名称":stores,"标的物":targets,"距离":distances,"地址":addresss}  #构造字典
df=pd.DataFrame(dict)
df.to_excel("检索结果.xlsx") 
```
